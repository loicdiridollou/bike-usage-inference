### Full worklow

1. Take one of the file from the list of files
2. Process the columns and rename/drop what needs to be done
3. Push the lines by package of 100,000

```python
creds = yaml.full_load(open('creds.yaml', 'r'))
driver = creds['driver']
username = creds['username']
password = creds['password']
host = creds['host']
port = creds.get('port', 0)
database = creds['database']

engine = sq.create_engine(
    '{driver}://{username}:{password}@{host}:{port}/{database}'.format(driver=driver,
                                                                       username=username,
                                                                       password=password,
                                                                       host=host,
                                                                       port=port,
                                                                       database=database)
)

for fname in list_files[1:-2]:
    print('Working on {}'.format(fname))
    df = pd.read_csv(s3_client.get_object(Bucket='gbfs-data', Key=fname)['Body'])
    if 'rideable_type' not in df:
        df['rideable_type'] = ''
    final_df = df.rename(columns=old_new).drop(columns=cols_to_drop, errors='ignore')
    final_df.to_sql(con=engine, name='bike_trips', if_exists='append', chunksize=50000)
```
