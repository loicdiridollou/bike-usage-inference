```python
import boto3
import pandas as pd
import json
import requests
```

```python
s3_client = boto3.client('s3')
dd = s3_client.list_objects_v2(Bucket='gbfs-data')
```

```python
list_files = []
for el in dd['Contents']:
    if el['Key'][-3:] == 'csv':
        list_files.append(el['Key'])
```

```python
list_files
```

```python
# get the station information
r = requests.get('https://gbfs.baywheels.com/gbfs/en/station_information.json')
response = r.json()['data']['stations']
```

```python
stations = pd.DataFrame(response)
```

```python
stations.head()
```

```python
list_columns = []
for fname in list_files:
    list_columns.append(pd.read_csv(s3_client.get_object(Bucket='gbfs-data', Key=fname)['Body']).columns)
```
